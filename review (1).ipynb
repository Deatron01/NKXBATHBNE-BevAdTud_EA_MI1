```python
# Task 1: Variables / 1. Feladat: Változók
# Create variables for a name and age. / Hozz létre változókat névhez és életkorhoz.
name = "Alice"
age = 25
print(f"Name: {name}, Age: {age}")
```

## 3. Pandas for Data Manipulation / Pandas az Adatmanipulációhoz
Pandas is used to clean and prepare data for analysis. / *A Pandast az adatok tisztítására és az elemzésre való előkészítésére használják.*

```python
import pandas as pd
# Task 10: DataFrame Creation / 10. Feladat: Adatkeret Létrehozása
data = {'Name': ['Anna', 'Béla', 'Anna'], 'Score': [10,20,30]}
df = pd.DataFrame(data)
print(df)
```

```python
# Task 11: Loading Data / 11. Feladat: Adatbetöltés
# Loading a CSV file into a DataFrame is a common task.
# Egy CSV fájl betöltése Adatkeretbe gyakori feladat.
df = pd.read_csv('sample_data/california_housing_test.csv')
```

```python
# Task 12: Missing Data / 12. Feladat: Hiányzó Adatok
# Working with missing data is part of preprocessing.
# A hiányzó adatok kezelése az előfeldolgozás része.
print(df.isnull().sum())
```

```python
# Task 13: Removing Duplicates / 13. Feladat: Duplikációk Eltávolítása
# Use drop_duplicates() to clean your dataset.
# Használd a drop_duplicates() függvényt az adatkészlet tisztításához.
df = df.drop_duplicates()
print(df)
```

```python
# Task 14: Grouping Data / 14. Feladat: Adatok Csoportosítása
# Aggregating and grouping data helps find trends.
# Az adatok összesítése és csoportosítása segít a trendek megtalálásában.
print(df.groupby('longitude').mean())
```

## 4. Analysis and Visualization / Elemzés és Vizualizáció
Visualizing data uses charts to interpret complex information. / *Az adatvizualizáció grafikonokat használ az összetett információk értelmezéséhez.*

```python
# Task 15: Correlation / 15. Feladat: Korreláció
# Calculating correlation helps find relationships between variables.
# A korreláció kiszámítása segít megtalálni a változók közötti kapcsolatokat.
print(df.corr(numeric_only=True))
```

```python
import matplotlib.pyplot as plt
# Task 16: Line Chart / 16. Feladat: Vonaldiagram
# Line charts are used for understanding trends.
# A vonaldiagramokat a trendek megértéséhez használják.
plt.plot(df['total_bedrooms'], df['median_income'])
plt.title("Trend Analysis")
plt.show()
```

```python
# Task 17: Histogram / 17. Feladat: Hisztogram
# Histograms show the distribution of data.
# A hisztogramok az adatok eloszlását mutatják.
plt.hist(df['housing_median_age'], bins=20)
plt.show()
```

```python
import seaborn as sns
# Task 18: Scatter Plot / 18. Feladat: Szórásdiagram
# Scatter plots help visualize bivariate analysis.
# A szórásdiagramok segítenek a kétváltozós elemzés vizualizálásában.
sns.scatterplot(x=df['housing_median_age'], y=df['total_rooms'])
plt.show()
```

## 5. Preprocessing for Machine Learning / Gépi Tanulás Előfeldolgozás
Scaling and encoding are vital for building accurate models. / *A skálázás és a kódolás alapvető fontosságú a pontos modellek építéséhez.*

```python
from sklearn.preprocessing import LabelEncoder
# Task 19: Label Encoding / 19. Feladat: Címke-kódolás
# Convert categorical data into numerical format.
# Konvertáld a kategóriás adatokat numerikus formátumba.
le = LabelEncoder()
df['total_rooms'] = le.fit_transform(df['total_rooms'])
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Task 20: Scaling / 20. Feladat: Skálázás
# Normalization transforms data into a specific range, like 0 to 1.
# A normalizálás meghatározott tartományba transzformálja az adatokat, például 0 és 1 közé.
scaler = MinMaxScaler()
scaled_scores = scaler.fit_transform(df[['population']])
print(scaled_scores)
```
